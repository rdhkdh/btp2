```python
import pandas as pd 
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Load CSV files into DataFrames
ue1_sinr = pd.read_csv('UE1_sinr.csv')
ue2_sinr = pd.read_csv('UE2_sinr.csv')
ue1_rssi = pd.read_csv('UE1_rssi.csv')
ue2_rssi = pd.read_csv('UE2_rssi.csv')
```

```python
ue1_sinr.head()
```

```python
ue1_rssi.head()
```

```python
# Plot SINR
plt.figure(figsize=(10, 5))
plt.plot(ue1_sinr.iloc[:, 0], ue1_sinr.iloc[:, 1], label='UE1 SINR')
plt.plot(ue2_sinr.iloc[:, 0], ue2_sinr.iloc[:, 1], label='UE2 SINR')
plt.xlabel('Time (seconds)')
plt.ylabel('SINR (dB)')
plt.title('SINR over Time')
plt.legend()
plt.grid()
plt.show()
```

```python
# Plot RSSI
plt.figure(figsize=(10, 5))
plt.plot(ue1_rssi.iloc[:, 0], ue1_rssi.iloc[:, 1], label='UE1 RSSI')
plt.plot(ue2_rssi.iloc[:, 0], ue2_rssi.iloc[:, 1], label='UE2 RSSI')
plt.xlabel('Time (seconds)')
plt.ylabel('RSSI (dBm)')
plt.title('RSSI over Time')
plt.legend()
plt.grid()
plt.show()
```

```python
# Determine 1/4th of the data
num_rows = len(ue1_rssi)  # Total number of rows
quarter_rows = num_rows // 500  # First quarter

# Plot RSSI for only the first quarter of the data
plt.figure(figsize=(10, 5))
plt.plot(ue1_rssi.iloc[:quarter_rows, 0], ue1_rssi.iloc[:quarter_rows, 1], label='UE1 RSSI')
plt.plot(ue2_rssi.iloc[:quarter_rows, 0], ue2_rssi.iloc[:quarter_rows, 1], label='UE2 RSSI')

plt.xlabel('Time (seconds)')
plt.ylabel('RSSI (dBm)')
plt.title('RSSI over Time (First 1/500th of Data)')
plt.legend()
plt.grid()
plt.show()
```
